## Mindestanforderungen

- Zombies kommen von einer Seite ins Feld in Zeilen. +
- Spieler positioniert Pflanzen. +
- Pflanzen schießen auf Zombies mit bestimmter Frequenz.
- Pflanzen und Zombies haben Lebenspunkte und werden zerstört, wenn diese aufgebraucht sind. 
- Wenn Zombie das Ziel erreicht, ist das Spiel verloren. +
- Wenn eine Mindestanzahl Zombies besiegt wurde, ist das Spiel gewonnen. +

Unsere Anforderungen:
- Grafiken einfuegen
- Sonnen
- Mehrere Pflanzen
***

## Bonuspunkte

Bonuspunkte kann es beispielsweise für folgende Erweiterungen geben:
- Verschiedene Pflanzenarten mit entsprechenden Effekten.
- Verschiedene Zombiearten mit entsprechenden Effekten.
- Kosten und Punkte.
- ...

Grundsätzlich kann es für alles Bonuspunkte geben, das die Spielfunktionalität erweitert oder das Spiel, über die Minimalanforderungen hinaus, näher an das Original heran bringt.

```c++
#include <algoviz/SVG.hpp>
#include <math.h>
#include <iostream>
#include <vector>
#include "Zombie.hpp"
#include "Plants.hpp"
#include "Kreaturen.hpp"

using namespace std;

SVG spiel = SVG(400,300);

//Hintergrund
Rect hintergrund = Rect( 0,0,400,300,0,0,&spiel);
hintergrund.setFill("lightgrey");

//Sonnenanzeige 
Rect sonnenAnzeige = Rect( 310, 10,80,30,0,0,&spiel);
sonnenAnzeige.setFill("white");
Image PicOfSun = Image("../Projekt/Pics/sun.png", 330, 25, 30, 30, &spiel);

Rect spielfeld[9][5];
int feld[9][5] = {0};

MouseState maus;

string taste = "";

vector<Plants> vektorPlant;
vector<Zombie> vektorZombie;
Zombie *pointer = nullptr;
Plants *pointer2 = nullptr;

//Erstellen der Spielfeldes
for ( int spalte = 0; spalte < 9; spalte = spalte+1 ) {
    for ( int zeile = 0 ; zeile < 5; zeile = zeile+1 ) {
        
        //Bewege das Rechteck an die richtige Stelle
        spielfeld[spalte][zeile].moveTo(spalte*30 + 130,zeile*40 + 75);
        
        //Setze es auf die richtige Größe
        spielfeld[spalte][zeile].setWidth(30);
        spielfeld[spalte][zeile].setHeight(40);
        
        //Farbe abwechselnd setzen
        if(spalte % 2 ^ zeile % 2){
            spielfeld[spalte][zeile].setFill("darkgreen");
        } else{
            spielfeld[spalte][zeile].setFill("green");
        }

        //Füge das Rechteck zur Zeichnung hinzu.
        spielfeld[spalte][zeile].addTo(&spiel);
    }    
}

//Erstellen des Pflanzen-Menus 
Rect plantsMenu[3];
for ( int spalte = 0; spalte < 3; spalte++) {
    
    //Rechteck an die richtige Stelle bewegen
    plantsMenu[spalte].moveTo(10, spalte*40 + 75);
    
    //Setze Groesse
    plantsMenu[spalte].setWidth(40);
    plantsMenu[spalte].setHeight(40);
    
    //Der Zeichnung hinzufuegen
    plantsMenu[spalte].addTo(&spiel);
    
    //Farbe auf weiss setzen
    plantsMenu[spalte].setFill("white");
}
//Bilder ins Pflanzenmenu einfuegen
Image PicOfPea = Image("../Projekt/Pics/pea-shooter.png", 30, 95, 30, 30, &spiel);

//Moegliche y-Koordinaten fuer die Zombies
int Y[5] = {95, 135, 175, 215, 255};
int y;
int zombieZaehler;
int z = 0;
int x = 400;

bool endWhile = false;
bool stehen = false;

while((taste != "x") &&  !endWhile){
    
    taste = spiel.lastKey();
    maus = spiel.lastClick();
    
    AlgoViz::sleep(150);
    
    //Frequenz Zombies
    if(z % 170 == 0 && zombieZaehler < 5){
        //Zufaellige Y-Koordinate
        y = Y[rand() % 5];
        
        //Zombies in Vektor
        pointer = new Zombie(x, y, spiel);
        vektorZombie.push_back(*pointer);
        
        //Zaehler von den Zombies
        zombieZaehler++;
    }
    
    for(int i = 0; i < vektorZombie.size(); i++){
        //Zombies laufen lassen
        if(vektorZombie[i].getStehen() == false){
           (vektorZombie[i]).Schritt(spiel); 
             cout << "b";
        }
        //Schleife beenden, wenn der Zombie den Rand beruehrt
        if((vektorZombie[i]).getX() <= 130){
            endWhile = true;
            cout << "a";
        }
        
        for(int j = 0; j < vektorPlant.size(); j++){
             cout << "f";
            
            //Y-Koordinaten vergleichen
            if(((vektorZombie[i]).getY() - (vektorPlant[j]).getY()) <= 10){
                 cout << "c";
                
                //Pflanze aktiviert sich und schiesst
                //Schuesse in Vektor
                //pointer3 = new Schuss((vektorPlant[j]).getX(), (vektorPlant[j]).getY(), spiel);
                //vektorShot.push_back(*pointer3);
                //laufen(vektorShot);
                //vektorPlant[j].activateShot(spiel);
                
                if(((vektorZombie[i]).getX() - (vektorPlant[j]).getXShot()) <= 10){
                    (vektorZombie[i]).setLebenspunkte((vektorZombie[i]).getLebenspunkte() - 10);
                }
                if((vektorZombie[i]).getLebenspunkte() == 0){
                        vektorZombie.erase(vektorZombie.begin() + i);
                        vektorZombie[i].Delete();
                }
                
                //X-Koordinaten vergleichen
                if(((vektorZombie[i]).getX() - (vektorPlant[j]).getX()) <= 20){
                    cout << "d";
                    
                    //Leben von den Pflanzen abziehen
                    (vektorPlant[j]).setLebenspunkte((vektorPlant[j]).getLebenspunkte() - 10);
                    cout << (vektorPlant[j]).getLebenspunkte();
                    
                    //Zombie stehen lassen
                    vektorZombie[i].setStehen(true);
                    
                    //Pflanze sterben lassen
                    if((vektorPlant[j]).getLebenspunkte() == 0){
                        vektorPlant.erase(vektorPlant.begin() + j);
                        vektorPlant[j].Delete();
                        
                        //Zombie weiter laufen lassen
                        (vektorZombie[i]).setStehen(false);
                        
                    }
                }
            }
        }
    }
    
    
    if(maus.isLegal() && maus.x()> 130 && maus.y() > 75 && maus.y() < 276){
        
        //Koordinaten des Feldes herausfinden
        int spalte = (maus.x() - 130) / 30;
        int zeile = (maus.y() - 75) / 40; 
        
        //feld == 0 heisst unbesetzt
        if(feld[spalte][zeile] == 0) {
            
            //Feld besetzen
            feld[spalte][zeile] = 1;
            
            //Mittelpunkt des Rechteckes herausfinden
            int MittelpunktX = spalte * 30 + 145;
            int MittelpunktY = zeile * 40 + 95;
            
            //Pflanze erstellen und in einen Vektor packen            
            pointer2 = new Plants(MittelpunktX , MittelpunktY, spiel);
            vektorPlant.push_back(*pointer2);
        }
    }
    z++;
}

//Als Schriftzug
if(endWhile == true){
    cout << "Sie haben leider verloren!";
}else if(taste == "x"){
    cout << "Das Spiel wurde beendet!";
}else{
    cout << "Sie haben gewonnen!";
}

//Alle Pflanzen und Zombies loeschen und uebergang zu Level 2
```
